```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data_to_load_df = pd.read_csv(city_data_to_load)
ride_data_to_load_df = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
combined_city_df = pd.merge(city_data_to_load_df, ride_data_to_load_df,
                                 how='outer', on='city')
```

```python
combined_city_df.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
#First aggregate # of rides per city
city = combined_city_df["city"]
fare = combined_city_df["fare"]
type = combined_city_df["type"]
date = combined_city_df["date"]
rides = combined_city_df["ride_id"]
#drivers = combined_city_df["driver_count"]



frame_df = pd.DataFrame({
    "city": [city],
    "fare": [fare],
    "type": [type],
    "date": [date]

    
    
    
    
})
frame_df

grouped_frame_df = combined_city_df.groupby(['city'])
#grouped_frame_df = combined_city_df.groupby(['type', 'city'])
#grouped_frame_df = grouped_frame_df.agg({"ride_id": "count"})
#grouped_frame_df = grouped_frame_df.agg({"city": "count", "ride_id": "count", "fare": "mean", "type": "count", "date": "count"})
grouped_frame_df.head()
```

```python
grouped_frame_df_city = grouped_frame_df.agg({"ride_id": "count", "fare": "mean", "driver_count": "mean"})
grouped_frame_df_city
```

```python
grouped_frame_df_type = combined_city_df.groupby(["city", "type", "driver_count"])
grouped_frame_df_type = grouped_frame_df_type.agg({"ride_id": "count", "fare": "mean"}).reset_index()
grouped_frame_df_type

```

```python
#make the number of rides per city your x coordinate and average fare $ your y coordinate
x_axis = grouped_frame_df_type[grouped_frame_df_type["type"]=="Urban"]["ride_id"]
y_axis = grouped_frame_df_type[grouped_frame_df_type["type"]=="Urban"]["fare"]
size_value = grouped_frame_df_type[grouped_frame_df_type["type"]=="Urban"]["driver_count"]
trace0 = plt.scatter(
     x= x_axis,
     y= y_axis,
     marker= "o",
     s= size_value*10,
     label= "Urban", 
     c= "orange",edgecolor= "black" 
        
     )


data = [trace0]

x_axis = grouped_frame_df_type[grouped_frame_df_type["type"]=="Rural"]["ride_id"]
y_axis = grouped_frame_df_type[grouped_frame_df_type["type"]=="Rural"]["fare"]
size_value = grouped_frame_df_type[grouped_frame_df_type["type"]=="Rural"]["driver_count"]
trace1 = plt.scatter(
     x= x_axis,
     y= y_axis,
     marker= "o",
     s= size_value*10,
     label= "Rural", 
     c= "yellow",edgecolor= "black"
        
     )


data = [trace1]

x_axis = grouped_frame_df_type[grouped_frame_df_type["type"]=="Suburban"]["ride_id"]
y_axis = grouped_frame_df_type[grouped_frame_df_type["type"]=="Suburban"]["fare"]
size_value = grouped_frame_df_type[grouped_frame_df_type["type"]=="Suburban"]["driver_count"]
trace2 = plt.scatter(
     x= x_axis,
     y= y_axis,
     marker= "o",
     s= size_value*10,
     label= "Suburban", 
     c= "lightskyblue", edgecolor= "black"
    )
data = [trace2]
#py.iplot(data, filename='bubblechart-size')
plt.xlabel ("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.legend(title= "City Type")
plt.grid(True)
plt.title("Pyber Ride Sharing Data (2016)")
plt.text(42,35,"Note: Circle size correlates with driver count per city")
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
Total_Fares = combined_city_df["fare"].sum()
Total_Fares

grouped_frame_df_CityType = combined_city_df.groupby(["type"])
grouped_frame_df_CityType = grouped_frame_df_CityType.agg({"fare": "sum"})
grouped_frame_df_CityType
```

```python
## Save Figure
# Labels for the sections of our pie chart
labels = ["Rural", "Suburban", "Urban"]

# The values of each section of the pie chart
sizes = [4328, 19356, 39854]

# The colors of each section of the pie chart
colors = ["gold", "lightskyblue", "lightcoral"]

# Tells matplotlib to seperate the "Python" section from the others
explode = (0, 0, 0.1)

plt.title("% of Total Fares by City Type")

# Creates the pie chart based upon the values above
# Automatically finds the percentages of each part of the pie chart
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)
```
